# Markov chain for a bacterial culture

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches

current_state = np.array([[1,0]])
transition_matrix = np.array([[0.45, 0.55], [0.023, 0.977]])

generations = [current_state[0][0]]
for i in range(20):
    next_state = current_state.dot(transition_matrix)
    generations.append(next_state[0][0])
    current_state = next_state
    
plt.plot(generations)
plt.xlim(0, 20)
plt.ylim(0, 1)
plt.grid()
plt.fill_between(range(21),generations, color='#539ecd')
plt.fill_between(range(21), generations, np.max(generations), color='#ffa500')
plt.title('Bacterial Culture', fontsize=20)

with_mutation = patches.Patch(color='#539ecd', label="with the mutation")
without_mutation = patches.Patch(color='#ffa500', label="without the mutation")
plt.legend(handles=[without_mutation, with_mutation])

plt.xlabel('generations')
plt.ylabel('percentage')
```

```python
print('After 20 generations, %.2f%% still have the mutation.' % (generations[-1] * 100))
```

# Stochastic Gradient Descent on Rosenbrock's function

```python
import numpy as np
import matplotlib.pyplot as plt
import random



def rosenbrock_f(x):
    return sum(100 * (x[1:] - x[:-1]**2)**2 + (x[:-1] - 1)**2)


def rosenbrock_df(x):
    gradient = np.zeros_like(x)
    gradient[0] = -400 * x[0] * (x[1] - x[0])**2 + 2 * x[0] - 2
    gradient[1:-1] = 200 * (x[1:-1] - x[:-2]**2) - 400 * x[1:-1] * (x[2:] - x[1:-1]**2.0) + 2 * x[1:-1] - 2
    gradient[-1] = 200 * (x[-1] - x[-2]**2)
    return gradient


def minimize(x, lrate, iterations=10000):
    points = np.array([[]])
    for i in range(iterations):
        points = np.append(points, [rosenbrock_f(x[i])])
        x = np.insert(x, len(x), np.subtract(x[i], lrate*rosenbrock_df(x[i])), axis=0)
    return x, points

def plot_lrate(x1, lrate, num_of_points=50, iterations=10000):
    res, points = minimize(x1, lrate, iterations)
    print("X = " + str(res[len(res)-1]) + " Rosenbrock(X) = " + str(rosenbrock_f(res[len(res)-1])))
    plt.plot(points[:num_of_points])
    plt.grid()
```

### From a starting point, which is actually already very close to the actual minimum and a learning rate, which is not too big and not too small, we converge to the minimum very fast.

```python
x1 = np.array([20 * [3]])
plot_lrate(x1, 1e-8)
```

### If we choose a learning rate, which is too big, we actually diverge away from the minimum, because the steps we take are too large.

```python
plot_lrate(x1, 1e-3)
```

### Here the learning rate is very small, so it takes very long for the gradient descent algorithm to reach the minimum.

```python
plot_lrate(x1, 1e-20, num_of_points=10000)
```

### Here we have choosen another starting point which is further away from the minimum and show, that with the same learning rate as the first example, it takes naturally longer to reach the minimum.

```python
x1 = np.array([20*[30]])
plot_lrate(x1, 1e-8)
```
